```python
import gc
import json
import numpy as np
import pandas as pd

from datetime import datetime

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
```

```python
datadir = '../data'

train = pd.read_json(datadir + '/train.json')
val = pd.read_json(datadir + '/val.json')
test = pd.read_json(datadir + '/test.json')
songmeta = pd.read_json(datadir + '/song_meta.json')

print("The shape of train: ", train.shape)
print("The shape of val: ", val.shape)
print("The shape of test: ", test.shape)

train['dataset'] = 'train'
val['dataset'] = 'val'
test['dataset'] = 'test'

full = pd.concat([train, val, test], axis = 0, ignore_index = True)

del train, val, test
gc.collect()
```

```python
full.dtypes
```

## Train 데이터 & Val 데이터 & Test 데이터 분포 확인
### 수록곡 개수
- val과 test는 거의 동일하게 분포되어 있음
    - 5곡 미만, 10 ~ 25곡 많음. 
    - 95곡~100곡도 많음
- train은 val, test에 비해 수록곡이 많음
    - 100곡 이상인 플레이리스트만 사용해야 하나?
    - 플레이리스트 개수가 적은 얘들이 label 역할을 할 수 있을까?

```python
full['num_of_songs'] = full['songs'].apply(lambda x: len(x))

plt.figure(figsize = (8, 5))
for dataset in ['train', 'val', 'test']:
    sns.distplot(a = full.loc[full['dataset'] == dataset, 'num_of_songs'], bins=40, label = dataset)
plt.legend()
plt.xlim(full['num_of_songs'].min(), full['num_of_songs'].max())
plt.title("The distribution of the number of songs for each dataset")
plt.xlabel("The number of songs")
plt.ylabel("Density")
plt.tight_layout()
plt.show()
```

### 태그 개수
- val과 test 비슷하개 분포되어 있음
- train은 무조건 1개 이상의 태그를 포함하고 있음

```python
full['num_of_tags'] = full['tags'].apply(lambda x: len(x))

plt.figure(figsize = (8, 5))
for dataset in ['train', 'val', 'test']:
    sns.distplot(a = full.loc[full['dataset'] == dataset, 'num_of_tags'], label = dataset)
plt.legend()
plt.xlim(full['num_of_tags'].min(), full['num_of_tags'].max())
plt.title("The distribution of the number of tags for each dataset")
plt.xlabel("The number of tags")
plt.ylabel("Density")
plt.tight_layout()
plt.show()
```

### 플레이리스트 업데이트 날짜
- train, val, test에서 플레이리스트의 업데이트 날짜의 분포가 유사

```python
full['updt_date'] = pd.to_datetime(full['updt_date'])
# timedelta from 2004-11-16
full['delta_updt_date'] = (full['updt_date'] - full['updt_date'].min()).apply(lambda x: x.days) 

plt.figure(figsize = (8, 5))
for dataset in ['train', 'val', 'test']:
    sns.distplot(a = full.loc[full['dataset'] == dataset, 'delta_updt_date'], label = dataset)
plt.legend()
plt.xlim(full['delta_updt_date'].min(), full['delta_updt_date'].max())
plt.title("The distribution of timedelta from 2004-11-16")
plt.xlabel("The time delta")
plt.ylabel("Density")
plt.tight_layout()
plt.show()
```

### Like 개수
- train, val, test에서 like 개수의 분포가 유사

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15,5))

for dataset in ['train', 'val', 'test']:
    sns.distplot(a = full.loc[full['dataset'] == dataset, 'like_cnt'], label = dataset, ax=axes[0])
    sns.distplot(a = np.log1p(full.loc[full['dataset'] == dataset, 'like_cnt']), label = dataset, ax=axes[1])
    
for i in range(2):
    axes[i].legend()
    
axes[0].set_title('The distribution of the number of likes')
axes[1].set_title('The distribution of log of the number of likes')

plt.show()
```

### 노래의 발매 YYYYMM

```python
full_expanding_songs = full.explode('songs')
full_expanding_songs = full_expanding_songs.merge(right = songmeta[['id', 'issue_date']], left_on = 'songs', right_on = 'id', how='left')
# 6,422 playlists doesn't have any song, and so drop those playlists
full_expanding_songs = full_expanding_songs.dropna(axis = 0).reset_index(drop = True)
full_expanding_songs['issue_year'] = full_expanding_songs['issue_date'].apply(lambda x: int(str(x)[:4]) if x != 0 else 0)
full_expanding_songs['issue_month'] = full_expanding_songs['issue_date'].apply(lambda x: int(str(x)[4:6]) if x != 0 else 0)
full_expanding_songs['issue_year_month'] = full_expanding_songs['issue_date'].apply(lambda x: int(str(x)[:6]) if x != 0 else 0)
```

**issue year**

```python
plt.figure(figsize = (8, 5))
for dataset in ['train', 'val', 'test']:
    sns.distplot(a = full_expanding_songs.loc[(full_expanding_songs['dataset'] == dataset) & (full_expanding_songs['issue_year'] != 0), 'issue_year'], label = dataset)
plt.legend()
plt.title("The distribution of the issue years of songs for each dataset")
plt.xlabel("Issue year")
plt.ylabel("Density")
plt.tight_layout()
plt.show()
```

## 가설 검정
### 플레이리스트 업데이트 날짜 이후에 발매된 곡을 수록할 수 없다.
- No, 플레이리스트 업데이트 날짜 이후에 발매된 곡도 플레이리스트에 포함되는 경우가 있다.
- 원곡 발매일은 플레이리스트 업데이트 이전 날짜지만, 메타데이터의 발매 일자는 재발매일 혹은 한국 발매일로 기록된 경우가 있다.

```python
full_expanding_songs['updt_date'] = full_expanding_songs['updt_date'].apply(lambda x: int(datetime.strftime(x, format='%Y%m%d')))
```

### train, val, test의 플레이리스트들에 수록된 곡들과 songmeta 곡들의 포함관계 확인
**val과 test의 곡 중에 train에 없는 곡이 있는가?**
- 학습 단계에서, 33,949곡에 해당하는 노드들은 업데이트 되지 않을 것. `train`만 볼 경우 이 곡들이 어느 곡들과 함께 등장하는지 알 수 없기 때문에 latent vector를 알 수 없다.
- 예측 단계에서, 같은 플레이리스트에 수록된 다른 곡들을 이용해서 예측은 가능
- train, val, test를 모두 사용해서 학습하면 위 곡들의 latent vector는 구할 수 있긴 하다. 하지만 full dataset을 사용할 경우 어떤 문제가 생길 것인가?
    - 이 부분은 리더보드 점수를 통해 파악하자.

```python
songs_in_train = full_expanding_songs.loc[full_expanding_songs['dataset']=='train', 'songs'].unique()
songs_in_test = full_expanding_songs.loc[full_expanding_songs['dataset'].isin(['val','test']), 'songs'].unique()
print("The number of songs: ", len(set(songs_in_test) - set(songs_in_train)))
```

**val과 test에 있는 플레이리스트들 중에 수록된 모든 곡들이 train에 없는 경우가 있는가?**
- 즉, 예측 단계에서도 협업 필터링을 전혀 사용하지 못하는 경우

### 훈련 데이터 플레이리스트 랜덤하게 지워보자

```python
p = 0.5

train = full.loc[full['dataset'] == 'train'].copy()
train['masked_songs'] = train['songs'].apply(lambda x: np.random.choice(x, size=int(len(x)*p), replace=False))
train['num_of_masked_songs'] = train['masked_songs'].apply(lambda x: len(x))
```

```python
plt.figure(figsize = (8, 5))
for dataset in ['train', 'val', 'test']:
    if dataset == 'train':
        sns.distplot(a = train['num_of_masked_songs'], bins=40, label = dataset)
    else:
        sns.distplot(a = full.loc[full['dataset'] == dataset, 'num_of_songs'], bins=40, label = dataset)
plt.legend()
plt.xlim(full['num_of_songs'].min(), full['num_of_songs'].max())
plt.title("The distribution of the number of songs for each dataset")
plt.xlabel("The number of songs")
plt.ylabel("Density")
plt.tight_layout()
plt.show()
```

### 이제 랜덤하게 20%는 playlist 없애버리자.

```python
p = 0.81

train['masked_vanished_songs'] = train['masked_songs'].apply(lambda x: x if np.random.choice(a=[0,1], size = 1, p=[1-p, p])[0] == 1 else [])
train['num_of_masked_vanished_songs'] = train['masked_vanished_songs'].apply(lambda x: len(x))
```

```python
plt.figure(figsize = (8, 5))
for dataset in ['train', 'val', 'test']:
    if dataset == 'train':
        sns.distplot(a = train['num_of_masked_vanished_songs'], bins=40, label = dataset)
    else:
        sns.distplot(a = full.loc[full['dataset'] == dataset, 'num_of_songs'], bins=40, label = dataset)
plt.legend()
plt.xlim(full['num_of_songs'].min(), full['num_of_songs'].max())
plt.title("The distribution of the number of songs for each dataset")
plt.xlabel("The number of songs")
plt.ylabel("Density")
plt.tight_layout()
plt.show()
```

```python
print("train set")
count_df = pd.value_counts(train['num_of_masked_vanished_songs'] > 0).reset_index()
count_df.columns = ['is_value', 'count']
count_df['relative_frequency'] = count_df['count'] / count_df['count'].sum()
count_df
```

## My Validation Set 만들어보기

```python
datadir = '../data'

train = pd.read_json(datadir + '/train.json')
val = pd.read_json(datadir + '/val.json')
test = pd.read_json(datadir + '/test.json')
songmeta = pd.read_json(datadir + '/song_meta.json')

print("The shape of train: ", train.shape)
print("The shape of val: ", val.shape)
print("The shape of test: ", test.shape)

train['dataset'] = 'train'
val['dataset'] = 'val'
test['dataset'] = 'test'

full = pd.concat([train, val, test], axis = 0, ignore_index = True)

del train, val, test
gc.collect()
```

```python
indices = full[full['dataset'] == 'train'].index

train_indices, val_indices = train_test_split(indices, test_size = 0.2, random_state = 0)
full.loc[val_indices, 'dataset'] = 'my_val'

condition = (full['dataset'] == 'my_val')
full.loc[condition, 'songs'] = full.loc[condition, 'songs'].apply(lambda x: np.random.choice(x, size=int(len(x)*0.5), replace=False))
full.loc[condition, 'songs'] = full.loc[condition, 'songs'].apply(lambda x: x if np.random.choice([0, 1], 1, p = [.2, .8])[0] == 1 else [])
```

```python
full['num_of_songs'] = full['songs'].apply(lambda x: len(x))

plt.figure(figsize = (8, 5))
for dataset in ['my_val', 'val', 'test']:
    sns.distplot(a = full.loc[full['dataset'] == dataset, 'num_of_songs'], bins=40, label = dataset)
plt.legend()
plt.xlim(full['num_of_songs'].min(), full['num_of_songs'].max())
plt.title("The distribution of the number of songs for each dataset")
plt.xlabel("The number of songs")
plt.ylabel("Density")
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15,5))

for dataset in ['my_val', 'val', 'test']:
    sns.distplot(a = full.loc[full['dataset'] == dataset, 'like_cnt'], label = dataset, ax=axes[0])
    sns.distplot(a = np.log1p(full.loc[full['dataset'] == dataset, 'like_cnt']), bins = 20, label = dataset, ax=axes[1])
    
for i in range(2):
    axes[i].legend()
    
axes[0].set_title('The distribution of the number of likes')
axes[1].set_title('The distribution of log of the number of likes')

plt.show()
```
