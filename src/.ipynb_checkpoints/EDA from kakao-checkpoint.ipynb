# 곡 장르 코드 데이터: `genre_gn_all.json'
- 총 254개 장르코드 (30개 대분류 장르코드 및 224개 상세 장르코드)

```python
genre_gn_all = pd.read_json('../data/genre_gn_all.json', typ = 'series', encoding = 'utf-8')
genre_gn_all = genre_gn_all.reset_index()
genre_gn_all.columns = ['gnr_code', 'gnr_name']
genre_gn_all
```

## 대분류 장르코드

```python
gnr_code = genre_gn_all[genre_gn_all['gnr_code'].str[-2:] == '00']
gnr_code
```

## 상세 장르코드

```python
dtl_gnr_code = genre_gn_all[genre_gn_all['gnr_code'].str[-2:] != '00']
dtl_gnr_code.columns = ['dtl_gnr_code', 'dtl_gnr_name']
dtl_gnr_code
```

## 장르 코드 트리(Tree)

```python
# Making join_code column
gnr_code = gnr_code.assign(join_code = gnr_code['gnr_code'].str[0:4])
dtl_gnr_code = dtl_gnr_code.assign(join_code = dtl_gnr_code['dtl_gnr_code'].str[0:4])

# merging
gnr_code_tree = pd.merge(left = gnr_code, right = dtl_gnr_code, how = 'left', on = 'join_code')
gnr_code_tree = gnr_code_tree.drop('join_code', axis = 1)
```

## 곡/태그/플레이리스트명 중복 비율은?

```python
# 1. 곡 별 수록된 플레이리스트 개수 count 테이블 생성: song_plist_cnt
song_plist_cnt = plist_songs_map.groupby('songs')['id'].nunique().reset_index(name = 'mapping_plist_cnt')


# 2. 단일/중복 수록 구분: 곡 별 수록된 플레이리스트가 한 개면 '단일 수록', 두 개 이상이면 '중복 수록'
song_plist_cnt = song_plist_cnt.assign(
    mapping_plist_cnt_category = pd.cut(song_plist_cnt['mapping_plist_cnt'], [0, 1, np.inf], labels = ['단일 수록', '중복 수록'])
)

# 3. 단일/중복 수록 구분 별 곡 수 count 테이블 생성: song_plist_mapping_division
song_plist_mapping_division = pd.DataFrame(song_plist_cnt.groupby('mapping_plist_cnt_category').count())

# 4. 3번 테이블에 비율 값 추가
song_plist_mapping_division = song_plist_mapping_division.rename(columns = {'songs': '곡 수'})
song_plist_mapping_division['비율(%)'] = round(song_plist_mapping_division['곡 수']/sum(song_plist_mapping_division['곡 수'])*100,2)
song_plist_mapping_division = song_plist_mapping_division.reset_index().rename(columns = {'mapping_plist_cnt_category':'중복 수록 여부'})
song_plist_mapping_division[['중복 수록 여부', '곡 수', '비율(%)']]
```

### 중복 수록 곡 중 많이 수록된 횟수 기준 상위 30곡

```python
# 1. song_plist_cnt 테이블에서 중복 수록 곡 중 수록 횟수 기준 상위 30곡  필터링 테이블 생성: duplicate_song_top30
duplicate_song_top30 = song_plist_cnt[song_plist_cnt['mapping_plist_cnt_category'] == '중복 수록'].nlargest(30, 'mapping_plist_cnt')

# 2-1 join 목적으로 곡 아이디 integer to string 변환
duplicate_song_top30 = duplicate_song_top30.merge(song_meta[['id', 'song_name', 'artist_name_basket']],
                                                  how = 'left', left_on = 'songs', right_on = 'id')
duplicate_song_top30 = duplicate_song_top30[['songs', 'song_name', 'artist_name_basket', 'mapping_plist_cnt']]
duplicate_song_top30 = duplicate_song_top30.rename(columns = {'songs':'곡 아이디',
                                                              'song_name':'곡 명',
                                                              'artist_name_basket':'아티스트 명',
                                                              'mapping_plist_cnt':'수록 획수'})
duplicate_song_top30
```

### 단일 수록 곡 중 가장 많은 곡이 수록된 장르 상위 30개

```python
# 1. song_plist_cnt 테이블에서 단일 수록 곡만 필터링하여 테이블 생성: single_song
single_song = song_plist_cnt[song_plist_cnt['mapping_plist_cnt_category'] == '단일 수록']

single_song_gnr = pd.merge(single_song, song_gnr_map, how = 'left', left_on = 'songs', right_on = 'song_id')
single_song_gnr_cnt = single_song_gnr.groupby('gnr_code')['songs'].nunique().reset_index(name = 'song_cnt')
single_song_gnr_cnt = single_song_gnr_cnt.merge(genre_gn_all, how = 'left', on = 'gnr_code')[['gnr_code', 'gnr_name', 'song_cnt']]
single_song_gnr_cnt.columns = ['대분류 장르코드', '장르명', '수록 곡 수']
single_song_gnr_cnt.sort_values('수록 곡 수', ascending = False)
```

### 중복 매핑 태그 비중

```python
# 1.  태그 별 수록된 플레이리스트 개수 count 테이블 생성: tag_plist_cnt
tag_plist_cnt = plist_tags_map.groupby('tags')['id'].nunique().reset_index(name = 'mapping_plist_cnt')

# 2. 단일/중복 매핑 구분: 태그 별 매핑된 플레이리스트가 한 개면 '단일 매핑', 두 개 이상이면 '중복 매핑'
tag_plist_cnt['mapping_plist_cnt_category'] = tag_plist_cnt['mapping_plist_cnt'].apply(lambda x: '단일 매핑' if x == 1 else '중복 매핑')

# 3. 단일/중복 매핑 구분 별 태그 수 Count 테이블 생성: tag_plist_mapping_division
tag_plist_mapping_division = pd.value_counts(tag_plist_cnt['mapping_plist_cnt_category']).reset_index()
tag_plist_mapping_division.columns = ['중복 매핑 여부', '태그 수']
tag_plist_mapping_division['비율(%)'] = round(tag_plist_mapping_division['태그 수'] / sum(tag_plist_mapping_division['태그 수']) * 100, 2)
tag_plist_mapping_division
```

### 플레이리스트명 중복 비중

```python
# 1. 플레이리스트 아이디(id)와 플레이리스트 명(plylst_title) 추출
plist_title = train[['id', 'plylst_title']]

# 2. 플레이리스트 이름 별 플레이리스트 ID count 테이블 생성: plist_id_cnt
plist_id_cnt = pd.value_counts(plist_title['plylst_title']).reset_index()
plist_id_cnt.columns = ['plist', 'mapping_plist_cnt']

# 3. 플레이리스트명 중복 구분: 서로 다른 플레이르스트 아이디로 겹치는 플레이리스트 이름이 없으면 '중복 없음', 그 외 '중복 있음'
plist_id_cnt['mapping_plist_cnt_category'] = plist_id_cnt['mapping_plist_cnt'].apply(lambda x: '중복 없음' if x==1 else '중복 있음')

# 4. 중복 구분 별 플레이리스트 수 count 테이블 생성: plist_id_cnt_division
plist_id_cnt_division = pd.value_counts(plist_id_cnt['mapping_plist_cnt_category']).reset_index()
plist_id_cnt_division.columns = ['중복 이름 여부', '플레이리스트 아이디 수']
plist_id_cnt_division['비율(%)'] = round(plist_id_cnt_division['플레이리스트 아이디 수'] / sum(plist_id_cnt_division['플레이리스트 아이디 수']) * 100,2)
plist_id_cnt_division
```

### 플레이리스트 태그 별 수록곡
- 태그 별 플레이리스트 수록 곡 특징을 탐색하여, 플레이리스트의 곡 장르 성격을 추정해볼 수 있다.

```python
# 1. 플레이리스트-곡 테이블 + 플레이리스트 태그 테이블 조인
# 하나의 플레이리스트에 있는 (곡, 태그) 조합 모두 생김
plist_song_tag_map = pd.merge(left=plist_songs_map, right=plist_tags_map, on='id', how='left')

# 2. 1번 테이블 + 곡 장르 테이블 join
plist_song_tag_gnr_map = pd.merge(left=plist_song_tag_map, right=song_gnr_map, left_on='songs', right_on='song_id', how='left')

# 3. 2번 테이블 + 장르 메타 테이블 join
gnr_code['gnr_code_name'] = gnr_code['gnr_code'] + ' (' + gnr_code['gnr_name'] + ')'
plist_song_tag_gnr_map = pd.merge(left=plist_song_tag_gnr_map, right=gnr_code[['gnr_code', 'gnr_code_name']], on='gnr_code', how='left')

# 4. 최종 테이블 필드 선택
plist_song_tag_gnr_map = plist_song_tag_gnr_map[['id', 'songs', 'tags', 'gnr_code_name']]
```

```python
# 태그 = ['드라이브', '스트레스', '기분전환']
sample1 = plist_song_tag_gnr_map.loc[plist_song_tag_gnr_map['tags'].isin(['드라이브', '기분전환', '스트레스'])]

# 태그 = ['밤', '새벽', '잔잔한']
sample2 = plist_song_tag_gnr_map.loc[plist_song_tag_gnr_map['tags'].isin(['밤', '새벽', '잔잔한'])]

# 태그 = ['사랑', '설렘']
sample3 = plist_song_tag_gnr_map.loc[plist_song_tag_gnr_map['tags'].isin(['사랑', '설렘'])]

# 태그 = ['추억', '회상']
sample4 = plist_song_tag_gnr_map.loc[plist_song_tag_gnr_map['tags'].isin(['추억', '회상'])]

# 태그 = ['슬픔', '이별']
sample5 = plist_song_tag_gnr_map.loc[plist_song_tag_gnr_map['tags'].isin(['슬픔', '이별'])]

# 태그 = ['휴식'. '힐링']
sample6 = plist_song_tag_gnr_map.loc[plist_song_tag_gnr_map['tags'].isin(['휴식', '힐링'])]

# 태그 = ['카페', '매장음악']
sample7 = plist_song_tag_gnr_map.loc[plist_song_tag_gnr_map['tags'].isin(['카페', '매장음악'])]
```

```python
# 각 샘플 태그 리스트 별 장르 비율 집계
sample1_gnr_rank = sample1.groupby('gnr_code_name').songs.nunique().reset_index(name='song_cnt')
sample1_gnr_rank['song_cnt_ratio'] = round(sample1_gnr_rank['song_cnt'] / sum(sample1_gnr_rank['song_cnt'])*100, 2)
sample1_gnr_rank['gnr_rank'] = sample1_gnr_rank['song_cnt_ratio'].rank(ascending=False)
sample1_gnr_rank['tag_list'] = '드라이브, 스트레스, 기분전환'
sample1_gnr_rank.columns = ['장르', '곡 수', '비율(%)', '순위', '태그 리스트']

# 각 샘플 태그 리스트 별 장르 비율 집계
sample2_gnr_rank = sample2.groupby('gnr_code_name').songs.nunique().reset_index(name='song_cnt')
sample2_gnr_rank['song_cnt_ratio'] = round(sample2_gnr_rank['song_cnt'] / sum(sample2_gnr_rank['song_cnt'])*100, 2)
sample2_gnr_rank['gnr_rank'] = sample2_gnr_rank['song_cnt_ratio'].rank(ascending=False)
sample2_gnr_rank['tag_list'] = '밤, 새벽, 잔잔한'
sample2_gnr_rank.columns = ['장르', '곡 수', '비율(%)', '순위', '태그 리스트']

# 각 샘플 태그 리스트 별 장르 비율 집계
sample3_gnr_rank = sample3.groupby('gnr_code_name').songs.nunique().reset_index(name='song_cnt')
sample3_gnr_rank['song_cnt_ratio'] = round(sample3_gnr_rank['song_cnt'] / sum(sample3_gnr_rank['song_cnt'])*100, 2)
sample3_gnr_rank['gnr_rank'] = sample3_gnr_rank['song_cnt_ratio'].rank(ascending=False)
sample3_gnr_rank['tag_list'] = '사랑, 설렘'
sample3_gnr_rank.columns = ['장르', '곡 수', '비율(%)', '순위', '태그 리스트']

# 각 샘플 태그 리스트 별 장르 비율 집계
sample4_gnr_rank = sample4.groupby('gnr_code_name').songs.nunique().reset_index(name='song_cnt')
sample4_gnr_rank['song_cnt_ratio'] = round(sample4_gnr_rank['song_cnt'] / sum(sample4_gnr_rank['song_cnt'])*100, 2)
sample4_gnr_rank['gnr_rank'] = sample4_gnr_rank['song_cnt_ratio'].rank(ascending=False)
sample4_gnr_rank['tag_list'] = '추억, 회상'
sample4_gnr_rank.columns = ['장르', '곡 수', '비율(%)', '순위', '태그 리스트']

# 각 샘플 태그 리스트 별 장르 비율 집계
sample5_gnr_rank = sample5.groupby('gnr_code_name').songs.nunique().reset_index(name='song_cnt')
sample5_gnr_rank['song_cnt_ratio'] = round(sample5_gnr_rank['song_cnt'] / sum(sample5_gnr_rank['song_cnt'])*100, 2)
sample5_gnr_rank['gnr_rank'] = sample5_gnr_rank['song_cnt_ratio'].rank(ascending=False)
sample5_gnr_rank['tag_list'] = '슬픔, 이별'
sample5_gnr_rank.columns = ['장르', '곡 수', '비율(%)', '순위', '태그 리스트']

# 각 샘플 태그 리스트 별 장르 비율 집계
sample6_gnr_rank = sample6.groupby('gnr_code_name').songs.nunique().reset_index(name='song_cnt')
sample6_gnr_rank['song_cnt_ratio'] = round(sample6_gnr_rank['song_cnt'] / sum(sample6_gnr_rank['song_cnt'])*100, 2)
sample6_gnr_rank['gnr_rank'] = sample6_gnr_rank['song_cnt_ratio'].rank(ascending=False)
sample6_gnr_rank['tag_list'] = '휴식, 힐링'
sample6_gnr_rank.columns = ['장르', '곡 수', '비율(%)', '순위', '태그 리스트']

# 각 샘플 태그 리스트 별 장르 비율 집계
sample7_gnr_rank = sample7.groupby('gnr_code_name').songs.nunique().reset_index(name='song_cnt')
sample7_gnr_rank['song_cnt_ratio'] = round(sample7_gnr_rank['song_cnt'] / sum(sample7_gnr_rank['song_cnt'])*100, 2)
sample7_gnr_rank['gnr_rank'] = sample7_gnr_rank['song_cnt_ratio'].rank(ascending=False)
sample7_gnr_rank['tag_list'] = '카페, 매장음악'
sample7_gnr_rank.columns = ['장르', '곡 수', '비율(%)', '순위', '태그 리스트']

del sample1, sample2, sample3, sample4, sample5, sample6, sample7
```

```python
# 행 방향으로 Concatenate 하기
samples_rbind = pd.concat(
    [sample1_gnr_rank, sample2_gnr_rank, sample3_gnr_rank,
     sample4_gnr_rank, sample5_gnr_rank, sample6_gnr_rank, sample7_gnr_rank],
    axis = 0
)

# 시각화
plt.figure(figsize = (20, 10))
sns.barplot(data=samples_rbind.loc[samples_rbind['순위']<=5,:],
            x='태그 리스트',
            y='비율(%)',
            hue='장르',
            edgecolor='black')
plt.title('각 태그 리스트 별 상위 장르 TOP 5 비교')
plt.xlabel('태그 리스트')
plt.ylabel('곡 비율 (%)')
plt.show()
```
